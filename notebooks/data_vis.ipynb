```python
import xarray as xr
```

## Monthly data

### Monthly data full

```python
file_name = "202001_mon_ERA5_full_ts.nc"
ds_full = xr.open_dataset(file_name)
ds_full
```

```python
ds_full["ts"].plot()
```

### Monthly data masked

```python
file_name = "202001_mon_ERA5_masked_ts.nc"
ds_masked = xr.open_dataset(file_name)
ds_masked
```

```python
ds_masked["ts"].plot()
```

## Daily data 

### Daily data full

```python
file_name = "202001_day_ERA5_full_ts.nc"
ds_full = xr.open_dataset(file_name)
ds_full
```

```python
ds_full["ts"].isel(time=0).plot()
```

### Daily data masked

```python
file_name = "202001_day_ERA5_masked_ts.nc"
ds_masked = xr.open_dataset(file_name)
ds_masked
```

```python
ds_masked["ts"].isel(time=0).plot()
```

```python
ds_masked["ts"].isel(lat=350, lon=70).plot()
```
